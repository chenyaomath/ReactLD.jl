```julia
# Parameters
const DIM = 3        # dim of the domain

p = (;
    Ω_inf = 0.0,       # nm
    Ω_sup = 200.0,     # nm
    T     = 0.25,      # s
    Δt    = 1.0e-6,    # s
    C₀    = 1.25e-4,   # nm^{-3}

    λ₁ = 90.3269519,   # 1/s, varies based on Ω
    λ₀ = 17.3,         # 1/s
    ε  = 10.0,         # nm
    γ₁ =  0.5,         # ratio
    γ₀ =  0.5,         # ratio

    D₁ = 1.0e6,        # nm 
    D₂ = 1.0e6,        # nm
    D₃ = 1.0e6,        # nm

    β₁_hat = 1.0,
    β₂_hat = 1.0
)
p = (; p..., 
    β₃_hat = (p.D₁ * p.β₁_hat * p.D₂ * p.β₂_hat) / (p.D₁ * p.β₁_hat + p.D₂ * p.β₂_hat) / p.D₃,
    D  = @SVector [p.D₁; p.D₂; p.D₃]
)
p = (; p...,
    β₁ = β * p.β₁_hat, 
    β₂ = β * p.β₂_hat,
    β₃ = β * p.β₃_hat,

    δ₁ = sqrt(2 * p.D₁ * p.Δt),
    δ₂ = sqrt(2 * p.D₂ * p.Δt),
    δ₃ = sqrt(2 * p.D₃ * p.Δt)
)
```

```julia
p = (; p..., 
    
    # diffusion
    β₁Δt_plus_1_inv = 1 / (p.β₁ * p.Δt + 1),
    β₂Δt_plus_1_inv = 1 / (p.β₂ * p.Δt + 1),
    β₃Δt_plus_1_inv = 1 / (p.β₃ * p.Δt + 1),

    β₁δ₁ = p.β₁ * p.δ₁,
    β₂δ₂ = p.β₂ * p.δ₂,
    β₃δ₃ = p.β₃ * p.δ₃,

    #reaction
    one_minus_γ₀ = 1 - p.γ₀,

    m1_over_m3 = (p.D₃ * p.β₃) / (p.D₁ * p.β₁),
    m2_over_m3 = (p.D₃ * p.β₃) / (p.D₂ * p.β₂), 
    
    sqrt_D_over_Δt = sqrt.(p.D / p.Δt),
    std_ζ = sqrt(p.D₁ * p.β₁ + p.D₂ * p.β₂),

    # update_state
    ε_square = (p.ε)^2,
    λ₁Δt = p.λ₁ * p.Δt,
    λ₀Δt = p.λ₀ * p.Δt
)
```

```julia
mutable struct Particle
    X::SVector{DIM, Float64}
    V::SVector{DIM, Float64}
end

mutable struct ReactState
    dir::Float64
    A::Particle
    B::Particle
    C::Particle
end
```

```julia
# Calculate the periodic distance between two positions
function period_dist_square(X1, X2, p)
    (; Ω_sup) = p
    diff1 = abs.(X1 - X2)
    diff2 = Ω_sup .- diff1
    return sum(min.(diff1, diff2) .^2)
end

# Test
test_a = @SVector [103.5;  48.6; 172.3]
test_b = @SVector [-54.2; 213.2; -22.9]

test_diff1 = abs.(test_a - test_b)
test_diff2 = p.Ω_sup .- abs.(test_a - test_b)
println("diff1:", test_diff1)
println("diff2:", test_diff2)
println("min{diff1, diff2}:", min.(test_diff1, test_diff2))
period_dist_square(test_a, test_b, p)
```

```julia
function forward_diff(X1, V1, X2, V2, p)
    # unpack variables from p 
    (; Ω_sup, Δt, β₁δ₁, β₂δ₂, β₁Δt_plus_1_inv, β₂Δt_plus_1_inv) = p

    Z1 = @SVector randn(3)
    V1_next = (V1 + β₁δ₁ * Z1) * β₁Δt_plus_1_inv
    X1_next = X1 + V1_next * Δt
    if !all(0.0 .<= X1_next .<= 200.0)
        X1_next = mod.(X1_next, Ω_sup)
    end

    Z2 = @SVector randn(3)
    V2_next = (V2 + β₂δ₂ * Z2) * β₂Δt_plus_1_inv    
    X2_next = X2 + V2_next * Δt
    if !all(0.0 .<= X2_next .<= 200.0)
        X2_next = mod.(X2_next, Ω_sup)
    end

    return (X1_next, V1_next,
            X2_next, V2_next)
end
```

```julia
# test
test_a = @SVector [103.5; 48.6; 172.3] # inside Ω
test_b = @SVector [-54.2; 213.2; -22.9] # outside Ω
test_V = @SVector rand(3)
test_V = 10 .* (2 .* test_V .- 1) .* p.sqrt_D_over_Δt

@benchmark forward_diff(test_a, test_V, test_a, test_V, p)
```

```julia
function backward_diff(X3, V3, p)
    (; Ω_sup, Δt, β₃δ₃, β₃Δt_plus_1_inv) = p

    # implicit method
    Z = @SVector randn(3)
    V3_next = (V3 + β₃δ₃ * Z) * β₃Δt_plus_1_inv
    X3_next = X3 + V3_next * Δt
    if all(0.0 .<= X3_next .<= 200.0)
        return (X3_next, V3_next)
    else
        return (mod.(X3_next, Ω_sup), V3_next)
    end
end

# test
test_a = @SVector [103.5;  48.6; 172.3] # inside Ω
test_b = @SVector [-54.2; 213.2; -22.9] # outside Ω
println("mod.(test_a, p.Ω_sup):", mod.(test_a, p.Ω_sup))
println("mod.(test_b, p.Ω_sup):", mod.(test_b, p.Ω_sup))

test_V = @SVector rand(3)
test_V = 10 .* (2 .* test_V .- 1) .* p.sqrt_D_over_Δt

backward_diff(test_a, test_V, p)
```

```julia
function forward_react(X1, V1, X2, V2, p)
    (; γ₁, Ω_sup, m1_over_m3, m2_over_m3) = p
    # X3 = γ₁ * X1 + (1 - γ₁) * X2
    X3 = X2 + γ₁ * (X1 - X2)
    if all(0.0 .<= X3 .<= 200.0)
        V3 = m1_over_m3 * V1 + m2_over_m3 * V2
        return (X3, V3)
    else
        V3 = m1_over_m3 * V1 + m2_over_m3 * V2
        return (mod.(X3, Ω_sup), V3)
    end
end

function backward_react(X3, V3, p)
    (; ε, γ₀, one_minus_γ₀, Ω_sup, std_ζ, m1_over_m3, m2_over_m3) = p

    # displacement of positions
    # sample uniform rv in B(0,ε)
    Z = @SVector randn(3)
    U = rand()
    η = ε * cbrt(U) * (Z / norm(Z))

    # original equation
    # γ₀ * X1 + (1-γ₀) * X2 = X3
    # X1 - X2 = η

    X1 = X3 + one_minus_γ₀ * η
    if !all(0.0 .<= X1 .<= 200.0)
        X1 = mod.(X1, Ω_sup)
    end
    X2 = X3 - γ₀ * η
    if !all(0.0 .<= X2 .<= 200.0)
        X2 = mod.(X2, Ω_sup)
    end

    # displacement of velocities
    ζ = std_ζ * @SVector randn(3)
    # original equation
    # (m₁/m₃) * V1 + (m₂/m₃) * V2 = V3
    # V1 - V2 = ζ
    V1 = V3 + m2_over_m3 * ζ
    V2 = V3 - m1_over_m3 * ζ

    return (X1, V1, X2, V2)
end

# test
test_X3 = @SVector [103.5; 48.6; 172.3]
test_V3 = @SVector rand(3)
test_V3 = 10 .* (2 .* test_V3 .- 1) .* p.sqrt_D_over_Δt

backward_react(test_X3, test_V3, p)
```

```julia
@benchmark backward_react($test_X3, $test_V3, $p)
```

```julia
function update_state(state, p)
    (; λ₁Δt, λ₀Δt, ε_square) = p

    # Forward reaction-diffusion 
    if state.dir == 1.0
        # forward diffusion
        X1, V1, X2, V2 = forward_diff(state.A.X, state.A.V,
                                      state.B.X, state.B.V, p)
        
        if (rand() <= λ₁Δt) && (period_dist_square(X1, X2, p) <= ε_square)
            # forward reaction
            X3, V3 = forward_react(X1, V1, X2, V2, p)
            state.C.X = X3
            state.C.V = V3
            state.dir = 0.0
        else
            state.A.X = X1
            state.A.V = V1
            state.B.X = X2
            state.B.V = V2
        end
    # Backward reaction-diffusion
    else
        # backward diffusion
        X3, V3 = backward_diff(state.C.X, state.C.V, p)

        if (rand() <= λ₀Δt)
            # backward reaction
            X1, V1, X2, V2 = backward_react(X3, V3, p)

            state.A.X = X1
            state.A.V = V1
            state.B.X = X2
            state.B.V = V2
            state.dir = 1.0
        else
            state.C.X = X3
            state.C.V = V3
        end
    end
end

# test
test_X1 = @SVector [172.3;  39.4; 109.5]
test_X2 = @SVector [ 98.2; 182.1;  91.6]
test_X3 = @SVector [ 12.3;  72.0;  24.8]

test_V1 = @SVector rand(3)
test_V1 = 10 .* (2 .* test_V3 .- 1) .* p.sqrt_D_over_Δt
test_V2 = @SVector rand(3)
test_V2 = 10 .* (2 .* test_V3 .- 1) .* p.sqrt_D_over_Δt
test_V3 = @SVector rand(3)
test_V3 = 10 .* (2 .* test_V3 .- 1) .* p.sqrt_D_over_Δt

test_A = Particle(test_X1, test_V1)
test_B = Particle(test_X2, test_V2)
test_C = Particle(test_X3, test_V3)
```

```julia
function init_states(num_states, p)
    (; Ω_sup, sqrt_D_over_Δt) = p
    
    react_states = Vector{ReactState}(undef, num_states)
    for i in 1:num_states
        init_dir = 0.0

        # init A
        init_X1 = Ω_sup * @SVector rand(3)
        init_V1 = @SVector rand(3)
        init_V1 = (20 .* init_V1 .- 10) .* sqrt_D_over_Δt
        init_A  = Particle(init_X1, init_V1)

        # init B
        init_X2 = Ω_sup * @SVector rand(3)
        init_V2 = @SVector rand(3)
        init_V2 = (20 .* init_V2 .- 10) .* sqrt_D_over_Δt
        init_B  = Particle(init_X2, init_V2)

        # init C
        init_X3 = Ω_sup * @SVector rand(3)
        init_V3 = @SVector rand(3)
        init_V3 = (20 .* init_V3 .- 10) .* sqrt_D_over_Δt
        init_C  = Particle(init_X3, init_V3)

        init_state = ReactState(init_dir, init_A, init_B, init_C)
        react_states[i] = init_state
    end
    return react_states
end
```

```julia
function count_C(react_states, num_states)
    num_AB = 0.0
    for j in 1:num_states
        num_AB += react_states[j].dir
    end
    return (num_states - num_AB)
end
```

```julia
# Test each simulation
# Δt = 1.0e-6

function test_each_sim(react_states, p)
    (; T, C₀, Ω_sup) = p

    Δt = 1.0e-6
    num_Δt     = Int(floor(T / Δt))  # 25000
    num_states = Int(C₀ * (Ω_sup)^3) # 1000

    save_time_step = T / 5000        # plot 5000 points
    num_save_steps = Int(floor(T / save_time_step)) + 1
    step_ratio     = Int(floor(save_time_step / Δt))
    num_C_sim_path = zeros(num_save_steps)

    num_C_sim_path[1] = Float64(num_states)

    for i in 1:num_Δt
        for j in 1:num_states
            update_state(react_states[j], p)
        end
        if mod(i, step_ratio) == 0
            num_C_sim_path[div(i, step_ratio)+1] = count_C(react_states, num_states)
        end
    end
end

num_states   = 1000
react_states = init_states(num_states, p)
test_each_sim(react_states, p)
```

```julia
ProfileView.@profview test_each_sim(react_states, p)
```

```julia
# Test outer loop
# Δt = 1.0e-7

function test_out_loop(num_sim, p)
    (; T, C₀, Ω_sup) = p

    Δt = 1.0e-7
    num_Δt     = Int(floor(T / Δt))  # 25000
    num_states = Int(C₀ * (Ω_sup)^3) # 1000

    save_time_step = T / 5000        # plot 5000 points
    num_save_steps = Int(floor(T / save_time_step)) + 1
    step_ratio     = Int(floor(save_time_step / Δt))
    num_C_sim_path = zeros(num_save_steps)
    num_C_total = zeros(num_save_steps)

    num_C_sim_path[1] = Float64(num_states)

    for _ in 1:num_sim
        react_states = init_states(num_states, p)
        for i in 1:num_Δt
            if mod(i, step_ratio) == 0
                num_C_sim_path[div(i, step_ratio)+1] = count_C(react_states, num_states)
            end
        end
        @. num_C_total += num_C_sim_path
    end
end

num_sim = 1000
test_out_loop(num_sim, p)
```

```julia
"""
    Main
"""

(; T, Δt, C₀, Ω_sup) = p

num_Δt     = Int(floor(T / Δt))
num_states = Int(C₀ * (Ω_sup)^3)

save_time_step = T / 5000 # plot 5000 points
step_ratio     = Int(floor(save_time_step / Δt))
num_save_steps = Int(floor(T / save_time_step)) + 1
num_C_sim_path = zeros(num_save_steps)
num_C_total    = zeros(num_save_steps)

num_C_sim_path[1] = Float64(num_states)

# Main Loop
@showprogress for n in 1:num_sim
    react_states = init_states(num_states, p)
    for i in 1:num_Δt
        for j in 1:num_states
            update_state(react_states[j], p)
        end
        if mod(i, step_ratio) == 0
            num_C_sim_path[div(i, step_ratio)+1] = count_C(react_states, num_states)
        end
    end
    @. num_C_total += num_C_sim_path
end

num_C_avg = num_C_total / num_sim
df = DataFrame(num_C=num_C_avg)
file_path = "test/AB_C_V_sim$num_sim.csv"
CSV.write(file_path, df)
```
